```python
import sys
import os
import pandas as pd
import numpy as np

from d2c.benchmark.cd_plot import draw_cd_diagram
```

The function expects a dataframe as follows 

| dataset_name   |      Model      |  Score |
|----------|:-------------:|------:|
| 1 |  d2c | 0.7 |
| 2 |    var   |  0.55 |
| 3 | dynoteas |    0.65 |


```python
import pandas as pd 
df = pd.read_csv('scores_no_filter_mixed_no_standardization.csv')
metrics = df.columns[5:10]
```

```python
df
```

```python
metrics
```

```python
for metric in metrics: 
    # we build one df per metric
    dfs_metric = []

    for method in df.method.unique():
        scores = df.loc[(df['method'] == method)][metric]
        scores = pd.DataFrame(scores)
        scores = scores.rename(columns={metric: 'Score'})
        scores["Model"] = method.upper()
        scores['dataset_name'] = np.arange(len(scores))

        dfs_metric.append(scores)

    df_metric = pd.concat(dfs_metric)

    path = f'cd_{metric}.png'

    mapping = {
        'accuracy': 'Accuracy',
        'balanced_accuracy': 'Balanced Accuracy',
        'precision': 'Precision',
        'recall': 'Recall',
        'f1': 'F1 Score'
    }
    draw_cd_diagram(path, df_perf=df_metric, title=mapping[metric], labels=False)
    
```
